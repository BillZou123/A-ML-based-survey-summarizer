```python
import pandas as pd
import re

df = pd.read_csv('project_data.csv', encoding = 'cp1252')
print(df.keys())
#data preprocess 1: replace nan with str "."
df = df.fillna(".")
#data preprocess 2: remove the answers in which both questions are not answered.
drop_list = []
for i in range(len(df['Intent2Use_Open'])):
  if (df['Intent2Use_Open'][i] == ".") and (df['Intentnotuse_Open'][i] == "."):
    drop_list.append(i)

df =df.drop(drop_list)
df = df.reset_index()
df
```

```python
#only keep the columns we want
df2 = df[['Intent2Use_Open','Intentnotuse_Open']]
```

```python
#output the preprocessed dataset to data.csv, this is the file we use for labelling.
df2.to_csv("data_preprocessed.csv")
```

```python
df = pd.read_csv("data_preprocessed.csv",index_col=[0])
def decontracted(phrase): 
    phrase = re.sub(r"won't", "will not", phrase) 
    phrase = re.sub(r"can\'t", "can not", phrase)  
    phrase = re.sub(r"n\'t", " not", phrase)  
    phrase = re.sub(r"\'re", " are", phrase)  
    phrase = re.sub(r"\'s", " is", phrase)  
    phrase = re.sub(r"\'d", " would", phrase)  
    phrase = re.sub(r"\'ll", " will", phrase)  
    phrase = re.sub(r"\'t", " not", phrase)  
    phrase = re.sub(r"\'ve", " have", phrase)  
    phrase = re.sub(r"\'m", " am", phrase)  
    return phrase
df['decontracted_Intent2Use_Open'] = df.apply(lambda row : decontracted(row['Intent2Use_Open']), axis = 1)
df['decontracted_Intentnotuse_Open'] = df.apply(lambda row : decontracted(row['Intentnotuse_Open']), axis = 1)
```

```python
# get precentage of intent/not intent to use
num_intent = 0
num_not_intent = 0
for i in range(len(df['Intent2Use_Open'])):
  if (df['Intent2Use_Open'][i]) == "." and df['Intentnotuse_Open'][i] != ".":
    num_not_intent +=1
  else:
    num_intent +=1

print(f"% intent: {round(num_intent/(num_intent + num_not_intent),4)*100}%")
print(f"% not intent: {round(num_not_intent/(num_intent + num_not_intent),4)*100}%")
```
