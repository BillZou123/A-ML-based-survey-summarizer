```python
import pandas as pd
from bertopic import BERTopic
```

```python
df = pd.read_csv("data.csv",index_col=[0])
df2 = df.loc[:,'decontracted_Intent2Use_Open']
```

```python
topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True,nr_topics=10)
topics, probs = topic_model.fit_transform(df2)
```

```python
freq = topic_model.get_topic_info()
names = freq['Name']
topic_names = []
for i in topics:
    if i != -1:
        topic_names.append(names[i+1])
    else:
        topic_names.append(names[0])
    
```

```python
freq['Name']
```

```python
topic_model.get_topic_info()
```

```python
df3 = pd.DataFrame(df2)
df3['topic_label']= topic_names
dic = {}
for i in freq['Name']:
    dic[i] = df3.loc[df3['topic_label'] == i ]['decontracted_Intent2Use_Open']
```

```python
final_df = pd.DataFrame(data=dic)
final_df.to_csv("data_with_topic.csv")
final_df
```
